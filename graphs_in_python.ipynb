```python
#pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(-3,3,20)
y = x**2
x
y
```

```python
plt.plot(x,y, '-b')
```

```python
plt.plot(x,y,'g*-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Title')
```

Multiplots on same canvas

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g*-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x is function of y')
plt.subplot(1,2,2)
plt.plot(y,x,'r*-')
plt.xlabel('y')
plt.ylabel('x')
plt.title('y is function of x')
```
